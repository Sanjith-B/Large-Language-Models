```python
import ast

def convert(text):
  list = []
  for i in ast.literal_eval(text):
    list.append(i['name'])

  return list

#example = df['genres'].apply(convert)

df['genres'] = df['genres'].apply(convert)
df['keywords'] = df['keywords'].apply(convert)
df['production_companies'] = df['production_companies'].apply(convert)
df['production_countries'] = df['production_countries'].apply(convert)
df['spoken_languages'] = df['spoken_languages'].apply(convert)
df['cast'] = df['cast'].apply(convert)

df.iloc[0]['crew']
```

```python
def fetch_director(text):
  list = []
  for i in ast.literal_eval(text):
    if i['job'] == 'Director':
      list.append(i['name'])
      break
  return list

def fetch_producer(text):
  list = []
  for i in ast.literal_eval(text):
    if i['job'] == 'Producer':
      list.append(i['name'])
      break
  return list

def fetch_writer(text):
  list = []
  for i in ast.literal_eval(text):
    if i['job'] == 'Writer':
      list.append(i['name'])
      break
  return list

df['Director'] = df['crew'].apply(fetch_director)
df['Producer'] = df['crew'].apply(fetch_producer)
df['Writer'] = df['crew'].apply(fetch_writer)

del df['crew']
```

```python
def parse_numeric_query(prompt):
    # Mapping from user phrases to numeric columns
    mapping = {
        "rating": "Ratings",
        "rated": "Ratings",
        "budget": "budget",
        "revenue": "revenue",
        "box office": "revenue",
        "time length": "runtime",
        "runtime": "runtime",
        "release" : "release_year"
        
    }

    # Look for descriptive terms
    conditions = {
        "highest": "max",
        "lowest": "min",
        "top": "max",
        "least": "min",
        "longest": "max",
        "shortest": "min",
        "latest" : "max",
        "newest" : "max",
        "older" : "min"
    }

    # Extract the column and condition
    selected_column = None
    condition = None

    for key, column in mapping.items():
        if re.search(key, prompt, re.IGNORECASE):
            selected_column = column
            break

    for key, cond in conditions.items():
        if re.search(key, prompt, re.IGNORECASE):
            condition = cond
            break

    return selected_column, condition
```

```python
def recommend_movies1(user_prompt, top_n=5):
    # Identify textual columns in the DataFrame
    textual_columns = df.select_dtypes(include=['object']).columns

    # Extract the name or keyword to search for in textual columns
    person_name = None
    match = re.search(r"of\s+(.+)", user_prompt, re.IGNORECASE)
    if match:
        person_name = match.group(1).strip()

    # Filter the DataFrame by searching for the name/keyword in all textual columns
    if person_name:
        mask = pd.DataFrame(
            {col: df[col].str.contains(person_name, case=False, na=False) for col in textual_columns}
        ).any(axis=1)
        filtered_df = df[mask]
        if filtered_df.empty:
            return f"No movies found for '{person_name}' in the dataset."
    else:
        # Textual similarity filtering
        cleaned_prompt = clean_text(user_prompt)
        query_embedding = model.encode([cleaned_prompt])[0]
        text_similarities = cosine_similarity([query_embedding], movie_embeddings)[0]
        top_text_indices = np.argsort(text_similarities)[::-1][:len(df) // 2]  # Take the top half as a filter
        filtered_df = df.iloc[top_text_indices]

    # Numeric query parsing
    numeric_column, condition = parse_numeric_query(user_prompt)

    if numeric_column and not filtered_df.empty:
        if condition == "max":
            # Sort movies by descending numeric column values
            sorted_indices = filtered_df[numeric_column].sort_values(ascending=False).index
        elif condition == "min":
            # Sort movies by ascending numeric column values
            sorted_indices = filtered_df[numeric_column].sort_values(ascending=True).index
        else:
            # If no specific condition, use mean for similarity
            target_value = filtered_df[numeric_column].mean()
            numeric_similarities = -np.abs(filtered_df[numeric_column] - target_value)
            # Normalize numeric similarities
            numeric_similarities = (numeric_similarities - numeric_similarities.min()) / (
                numeric_similarities.max() - numeric_similarities.min()
            )
            combined_similarities = 0.7 * text_similarities + 0.3 * numeric_similarities
            top_indices = np.argsort(combined_similarities)[::-1][:top_n]
            recommendations = filtered_df.iloc[top_indices][['title', 'Ratings', 'runtime', 'cast', 'Director','genres','release_year']]
            return recommendations
    else:
        numeric_similarities = np.ones(len(filtered_df))  # Equal weight to all

    # Combine textual and numeric similarities
    if numeric_column and condition:
        # Prioritize based on the sorted indices for max/min conditions
        sorted_filtered_df = filtered_df.loc[sorted_indices]
        recommendations = sorted_filtered_df[['title', 'Ratings', 'runtime', 'cast', 'Director','genres','release_year']].head(top_n)
    else:
        # Use combined similarities for default behavior
        combined_similarities = 0.7 * text_similarities + (
            0.3 * numeric_similarities if numeric_column else 0
        )
        top_indices = np.argsort(combined_similarities)[::-1][:top_n]
        recommendations = filtered_df.iloc[top_indices][['title', 'Ratings', 'runtime', 'cast', 'Director','genres','release_year']]

        if not numeric_column and not condition:
    # Default to top matches by text similarity for single-word prompts
          top_indices = np.argsort(text_similarities)[::-1][:top_n]
          recommendations = df.iloc[top_indices][['title', 'Ratings', 'runtime', 'cast', 'Director','genres','release_year']]



    return recommendations
```

```python
def process_user_prompt(user_prompt):

    match = re.search(r"of\s+(.+)", user_prompt, re.IGNORECASE)
    
    if match:

        return recommend_movies1(user_prompt)
    else:

        cleaned_prompt = clean_text(user_prompt)
        query_embedding = model.encode([cleaned_prompt])[0]

        # Calculate cosine similarities
        similarities = cosine_similarity([query_embedding], movie_embeddings)

        # Get top 5 matches
        top_indices = np.argsort(similarities[0])[::-1][:5]
        top_movies = df.iloc[top_indices][['title', 'Ratings', 'genres', 'cast', 'Director', 'runtime','release_year']]
        return top_movies


user_prompt = "Christian Bale"
result = process_user_prompt(user_prompt)
result
```

```python
user_history = []

def recommend_movies1(user_prompt, top_n=5):
    global user_history

    # Store the cleaned user prompt in history
    cleaned_prompt = clean_text(user_prompt)
    user_history.append(cleaned_prompt)

    # Identify textual columns in the DataFrame
    textual_columns = df.select_dtypes(include=['object']).columns

    # Extract the name or keyword to search for in textual columns
    person_name = None
    match = re.search(r"of\s+(.+)", user_prompt, re.IGNORECASE)
    if match:
        person_name = match.group(1).strip()

    # Filter the DataFrame by searching for the name/keyword in all textual columns
    if person_name:
        mask = pd.DataFrame(
            {col: df[col].str.contains(person_name, case=False, na=False) for col in textual_columns}
        ).any(axis=1)
        filtered_df = df[mask]
        if filtered_df.empty:
            return f"No movies found for '{person_name}' in the dataset.", None
    else:
        # Textual similarity filtering
        query_embedding = model.encode([cleaned_prompt])[0]
        text_similarities = cosine_similarity([query_embedding], movie_embeddings)[0]
        top_text_indices = np.argsort(text_similarities)[::-1][:len(df) // 2]  # Take the top half as a filter
        filtered_df = df.iloc[top_text_indices]

    # Generate direct recommendations
    query_embedding = model.encode([cleaned_prompt])[0]
    text_similarities = cosine_similarity([query_embedding], movie_embeddings)[0]
    top_indices = np.argsort(text_similarities)[::-1][:top_n]
    direct_recommendations = df.iloc[top_indices][['title', 'Ratings', 'runtime', 'cast', 'Director', 'genres', 'release_year']]

    # Generate history-based recommendations
    if len(user_history) > 5:
        history_embeddings = model.encode(user_history)
        avg_embedding = np.mean(history_embeddings, axis=0)
        history_similarities = cosine_similarity([avg_embedding], movie_embeddings)[0]
        history_top_indices = np.argsort(history_similarities)[::-1][:10]
        history_recommendations = df.iloc[history_top_indices][['title', 'Ratings', 'runtime', 'cast', 'Director', 'genres', 'release_year']]
    else:
        history_recommendations = None

    return direct_recommendations, history_recommendations

```
